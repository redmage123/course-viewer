### Exercise 1.3: User Input and Type Casting

**Important:** The `input()` function ALWAYS returns a string, even when the user types a number!

```python
# Exercise 1.3: User Input and Type Casting

# Step 1: Get input from user (for testing, we'll simulate this)
user_number = "25"  # In a real scenario: input("Enter a number: ")

# Step 2: Check the type - it's a string!
print(f"user_number = {user_number}")
print(f"Type: {type(user_number)}")

# Step 3: Convert to integer using int()
number_as_int = int(user_number)

# Step 4: Now do math with it
doubled = number_as_int * 2

print(f"After conversion: {number_as_int}, Type: {type(number_as_int)}")
print(f"Doubled: {doubled}")

# Bonus: Try other casting functions
# Convert "3.14" to a float
pi_string = "3.14"
pi_float = float(pi_string)

# Convert the number 42 to a string  
num = 42
num_string = str(num)

# Convert 0 and 1 to booleans
zero_bool = bool(0)
one_bool = bool(1)

print(f"\nBonus conversions:")
print(f"String '3.14' to float: {pi_float}")
print(f"Number 42 to string: {num_string}, Type: {type(num_string)}")
print(f"0 as bool: {zero_bool}, 1 as bool: {one_bool}")
```

## Part 2: Strings

### Exercise 2.1: String Basics

```python
message = "Python is amazing!"

# Exercise 2.1: String operations
length = len(message)
uppercase = message.upper()
first_five = message[:5]
last_seven = message[-7:]

print(f"Length: {length}")
print(f"Uppercase: {uppercase}")
print(f"First 5: {first_five}")
print(f"Last 7: {last_seven}")
```

```python
email = "john.doe@example.com"

# Exercise 2.2: String methods
has_at = "@" in email
parts = email.split("@")
new_email = email.replace("example", "gmail")

print(f"Contains @: {has_at}")
print(f"Username: {parts[0]}, Domain: {parts[1]}")
print(f"New email: {new_email}")
```

## Part 3: Lists

### Exercise 3.1: List Creation and Access

```python
# Exercise 3.1: Lists
foods = ["pizza", "sushi", "tacos", "pasta", "curry"]

first = foods[0]
last = foods[-1]
middle = foods[1:4]

print(f"First food: {first}")
print(f"Last food: {last}")
print(f"Middle foods: {middle}")
```

### Exercise 3.2: List Modification

```python
numbers = [1, 2, 3, 4, 5]

# Exercise 3.2: Modify the list
# Step 1: Add 6 to the end
numbers.append(6)

# Step 2: Insert 0 at the beginning
numbers.insert(0, 0)

# Step 3: Remove the value 3
numbers.remove(3)

# Step 4: Sort in reverse order
numbers.sort(reverse=True)

print(f"Final list: {numbers}")
# Expected: [6, 5, 4, 2, 1, 0]
```

### Exercise 3.3: List Comprehension

```python
words = ["python", "is", "awesome", "and", "fun"]

# Exercise 3.3: List comprehension
squares = [x ** 2 for x in range(1, 11)]
evens = [x for x in range(1, 21) if x % 2 == 0]
lengths = [len(word) for word in words]

print(f"Squares: {squares}")
print(f"Evens: {evens}")
print(f"Word lengths: {lengths}")
```

## Part 4: Dictionaries

### Exercise 4.1: Create a Dictionary

```python
# Exercise 4.1: Create a dictionary
book = {
    "title": "The Hitchhiker's Guide to the Galaxy",
    "author": "Douglas Adams",
    "year": 1979,
    "pages": 224,
    "is_fiction": True
}

print(f"Book: {book['title']} by {book['author']}")
print(f"Published: {book['year']}, Pages: {book['pages']}")
```

### Exercise 4.2: Dictionary Operations

```python
# Exercise 4.2: Dictionary operations

# Step 1: Add rating
book["rating"] = 4.5

# Step 2: Update pages (add 10)
book["pages"] = book["pages"] + 10
# Or: book["pages"] += 10

# Step 3: Get author safely
author = book.get("author")

# Step 4: Get keys and values
all_keys = book.keys()
all_values = book.values()

print(f"Updated book: {book}")
print(f"Keys: {list(all_keys)}")
print(f"Values: {list(all_values)}")
```

## Part 5: Control Flow

### Exercise 5.1: If Statements

```python
# Exercise 5.1: If statements
number = -5
person_age = 17

# Check if number is positive, negative, or zero
if number > 0:
    print(f"{number} is positive")
elif number < 0:
    print(f"{number} is negative")
else:
    print(f"{number} is zero")

# Check if person can vote
if person_age >= 18:
    print(f"At age {person_age}, you can vote!")
else:
    print(f"At age {person_age}, you cannot vote yet. Wait {18 - person_age} more year(s).")
```

### Exercise 5.2: For Loops

```python
# Exercise 5.2: For loops

# 1. Print each character with index
print("Characters in 'Python':")
for i, char in enumerate("Python"):
    print(f"  Index {i}: '{char}'")

# 2. Sum of 1 to 100
total = 0
for i in range(1, 101):
    total += i
print(f"\nSum of 1 to 100: {total}")

# 3. Numbers divisible by 3
divisible_by_3 = []
for i in range(1, 31):
    if i % 3 == 0:
        divisible_by_3.append(i)
# Or using list comprehension: divisible_by_3 = [i for i in range(1, 31) if i % 3 == 0]
print(f"Divisible by 3: {divisible_by_3}")
```

### Exercise 5.3: While Loop

```python
# Exercise 5.3: While loop countdown
count = 10

while count >= 1:
    print(count)
    count -= 1

print("Liftoff!")
```

## Part 6: Functions

### Exercise 6.1: Simple Function

```python
# Exercise 6.1: Simple function
def greet(name):
    return f"Hello, {name}!"

# Test
print(greet("Alice"))
print(greet("Bob"))
```

### Exercise 6.2: Function with Default Parameter

```python
# Exercise 6.2: Function with default parameter
def power(base, exponent=2):
    return base ** exponent

# Test
print(power(5))      # 25 (5^2)
print(power(2, 3))   # 8 (2^3)
print(power(10, 0))  # 1 (10^0)
```

### Exercise 6.3: Function with Multiple Returns

```python
# Exercise 6.3: Function with multiple returns
def analyze_list(numbers):
    minimum = min(numbers)
    maximum = max(numbers)
    average = sum(numbers) / len(numbers)
    return minimum, maximum, average

# Test
test_numbers = [4, 8, 15, 16, 23, 42]
minimum, maximum, average = analyze_list(test_numbers)
print(f"Min: {minimum}, Max: {maximum}, Avg: {average}")
```
