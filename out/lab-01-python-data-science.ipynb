## 1.3 Reshaping Arrays

| Example | Description |

## 4.1 Normalization

Normalization scales values to the range [0, 1] using the formula:


In NumPy: `(arr - arr.min()) / (arr.max() - arr.min())`

## 4.2 Standardization

Standardization centers data around 0 with standard deviation of 1:
